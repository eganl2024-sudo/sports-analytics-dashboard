```python
# Load Premier League Data using MatchHistory (Results & Odds)
mh = sd.MatchHistory(leagues="ENG-Premier League", seasons=2022)

pl_results = mh.read_games()

print(pl_results.head())

# Load Premier League Data using Understat
us = sd.Understat(leagues="ENG-Premier League", seasons=2022)

pl_understat = us.read_schedule()

print(pl_understat.head())

```

```python
# Load ClubElo Ratings
# ClubElo provides Elo ratings for clubs.
print("--- ClubElo Methods ---")
print([m for m in dir(sd.ClubElo) if m.startswith('read_')])

# Read ClubElo ratings by date
elo = sd.ClubElo() 

elo_ratings = elo.read_by_date("2023-05-28")
print("ClubElo Ratings (2023-05-28):")
print(elo_ratings.head())

# Read ClubElo ratings by team
elo_ManU = elo.read_team_history(team ="ManUnited")

print(elo_ManU.head())

# Try another team
elo_Chelsea = elo.read_team_history(team="Chelsea")
print(elo_Chelsea.head())
```

```python
# Preview elo by date columns
print("ClubElo Columns:")
print(elo_ratings.columns.tolist())

# Preview elo by team columns
print("ClubElo Team History Columns:")
print(elo_ManU.columns.tolist())
```

```python
# Preview: Manchester United Elo history
# Note: Team names must match ClubElo's format (usually "Man United")
# Filtering from 2015-01-01 to present
man_utd_elo = elo.read_team_history("Man United")

# Filter for dates after 2015-01-01
man_utd_recent = man_utd_elo[man_utd_elo.index >= "2015-01-01"]

print("Man United Elo History (Since 2015):")
print(man_utd_recent.head())
print("\nRecent Ratings:")
print(man_utd_recent.tail())

# Simple Plot to visualize
man_utd_recent['elo'].plot(title="Man United Elo Rating (2015-Present)", figsize=(10, 5))
```

```python
# Repeat for Real Madrid to verify consistency across leagues
real_madrid_elo = elo.read_team_history("Real Madrid")
real_madrid_recent = real_madrid_elo[real_madrid_elo.index >= "2015-01-01"]
print("Real Madrid Elo History (Since 2015):")
print(real_madrid_recent.head())
print("\nRecent Ratings:")
print(real_madrid_recent.tail())

# Simple Plot to visualize
real_madrid_recent['elo'].plot(title="Real Madrid Elo Rating (2015-Present)", figsize=(10, 5))
```
